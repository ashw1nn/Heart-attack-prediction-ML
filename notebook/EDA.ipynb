# Encoding

**Columns to OneHotEncode are cp, restecg, slp, caa, thall**

```python
cp = pd.get_dummies(dataset["cp"])
cp.columns = ["cp0", "cp1", "cp2", "cp3"]
restecg = pd.get_dummies(dataset["restecg"])
restecg.columns = ["recg0", "recg1", "recg2"]
slp = pd.get_dummies(dataset["slp"])
slp.columns = ["slp0", "slp1", "slp2"]
caa = pd.get_dummies(dataset["caa"])
caa.columns = ["caa0", "caa1", "caa2", "caa3", "caa4"]
thall = pd.get_dummies(dataset["thall"])
thall.columns = ["thall0", "thall1", "thall2", "thall3"]
dataset_final = pd.concat([dataset, cp, restecg, slp, caa, thall], axis=1)
dataset_final.drop(labels=["cp", "restecg", "slp", "caa", "thall"], axis=1, inplace=True)
dataset_final.head()
```
